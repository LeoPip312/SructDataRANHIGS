## Задание 1:
### Разработайте алгоритм, который запрашивает у пользователя четыре целых числа и затем определяет, есть ли среди введенных чисел одинаковые или нет. Нарисуйте блок-схему алгоритма.

```python
len_numlist = 4
flag_dublicate = 0

print("Введите 4 целых числа")

# Проверка на "дурака" проводиться будет!
my_numlist = [int(input()) for i in range(len_numlist)]
print("Введённые числа - ", my_numlist)


for i in range(len_numlist):
    for j in range(i + 1, len_numlist):
        if my_numlist[i] == my_numlist[j]:
            flag_dublicate = 1
            break

if flag_dublicate:
    print("Числа в списке повторяются!")
else:
    print("Числа в списке не повторяются!") 
```

## Задание 2:
### Разработайте алгоритм, который запрашивает у пользователя четыре целых числа и затем находит третье по величине число, если оно существует. Нарисуйте блок-схему алгоритма.

```python
# Пузырьковая сортировка по убыванию
def sort(arr):
    for i in range(len(arr) - 1):
        for j in range(0, len(arr) - i - 1):
            if arr[j] < arr[j+1]:
                temp = arr[j] 
                arr[j] = arr[j+1]
                arr[j+1] = temp
                
# Поиск 3-его максимального числа
def find_3_max(arr):
    max_val = arr[0]
    max_counter = 1;
    
    for i in range(1, len(arr)):
        if max_val > arr[i]:
            max_val = arr[i]
            max_counter += 1
        
        if max_counter == 3:
            return max_val
        
    return None
```

```python
len_numlist_2 = 4

print("Введите 4 целых числа")

# Проверка на "дурака" проводиться будет!
my_numlist_2 = [int(input()) for i in range(len_numlist_2)]
print("Введённые числа - ", my_numlist_2)

sort(my_numlist_2)
print("Отсортированные числа - ", my_numlist_2)

max_3 = find_3_max(my_numlist_2)
if max_3 != None:
    print("Третье по величине число - ", max_3)
else:
    print("Третьего по величине числа не существует")
```

## Задание 3:
### Разработайте алгоритм, который случайным образом задает четыре целых числа и затем определяет максимальное и минимальное число, а также находит количество максимальных и минимальных чисел среди введенных. Нарисуйте блок-схему алгоритма.

```python
# Импорт библиотеки для рандомных чисел
from random import randint
```

```python
# Функция поиска максимального и минимального элемента, а так же подсчёт количества максимальных и минимальных элементов
def find_max_min_values(arr):
    
    min_value = max_value = arr[0]
    count_max = count_min = 0
    
    for i in range(len(arr)):
        
        if min_value > arr[i]:
            min_value = arr[i]
        
        if max_value < arr[i]:
            max_value = arr[i]
        
    for i in range(len(arr)):
        
        if max_value == arr[i]:
            count_max += 1
            
        if min_value == arr[i]:
            count_min += 1
        
    return max_value, min_value, count_max, count_min
```

```python
len_numlist_3 = 4

my_numlist_3 = [randint(0,9) for i in range(len_numlist_3)]
print("Полученный массив - ", my_numlist_3)

result = find_max_min_values(my_numlist_3)
print(f"Максимальное число - {result[0]}, количество максимальных чисел - {result[2]}")
print(f"Минимальное число - {result[1]}, количество минимальных чисел - {result[3]}")
```
